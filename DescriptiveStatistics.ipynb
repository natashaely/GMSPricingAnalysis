### **Importing the libraries and reading in the data**

```python
#Importing the libraries required for this project.

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
```

```python
#Reading in the data.

df = pd.read_csv('/content/drive/MyDrive/Colab/7 GMS Pricing Analysis/AllProducts.csv')
```



---



###  **Exploring the data**

```python
#Reviewing the data.

df.head()
```

```python
#Reviewing the columns for any missing data.

for col in df.columns:
  pct_missing = np.mean(df[col].isnull())
  print('{} - {}%'.format(col, pct_missing))
```



---



###  **Descriptive Analysis**

```python
# Calculate summary statistics for price_in_usd and total_sold_quantity
summary_stats = df[['price_in_usd', 'total_sold_quantity']].describe()
summary_stats
```



---



### **Visualising the data: Histograms and Boxplots**

```python
# Set style for seaborn plots
sns.set_style("whitegrid")

# Initialize the figure
plt.figure(figsize=(15, 10))

# Create subplots for histograms, boxplots, and scatter plot
ax1 = plt.subplot(2, 2, 1)
ax2 = plt.subplot(2, 2, 2)
ax3 = plt.subplot(2, 2, 3)
ax4 = plt.subplot(2, 2, 4)

# Histogram for price_in_usd
sns.histplot(df['price_in_usd'], ax=ax1, bins=30, color='blue')
ax1.set_title('Histogram of Prices')
ax1.set_xlabel('Price in USD')
ax1.set_ylabel('Frequency')

# Histogram for total_sold_quantity
sns.histplot(df['total_sold_quantity'], ax=ax2, bins=30, color='red')
ax2.set_title('Histogram of Total Sold Quantity')
ax2.set_xlabel('Total Sold Quantity')
ax2.set_ylabel('Frequency')

# Box plot for price_in_usd
sns.boxplot(y=df['price_in_usd'], ax=ax3, color='blue')
ax3.set_title('Boxplot of Prices')
ax3.set_ylabel('Price in USD')

# Box plot for total_sold_quantity
sns.boxplot(y=df['total_sold_quantity'], ax=ax4, color='red')
ax4.set_title('Boxplot of Total Sold Quantity')
ax4.set_ylabel('Total Sold Quantity')

# Adjust layout
plt.tight_layout()
plt.show()
```



---



### **Visualising the data: Scatterplot**

```python
# Scatter plot between price_in_usd and total_sold_quantity

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['price_in_usd'], y=df['total_sold_quantity'], alpha=0.6)
plt.title('Scatter Plot of Price vs. Total Sold Quantity')
plt.xlabel('Price in USD')
plt.ylabel('Total Sold Quantity')
plt.show()
```
